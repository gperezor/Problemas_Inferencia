```python
#______Problema 1________

# a) tener en cuenta que 1 pulgada = 2.54 cm

df = pd.read_excel('VEF.xls')
height_cm = df['Altura(pulgadas)']*2.54
df['Altura(pulgadas)']=height_cm
df.rename(columns={'Altura(pulgadas)': 'Altura(cm)'}, inplace=True)
df
```

```python
df2 = df[['Edad','VEF','Altura(cm)']]
df2.describe()
```

```python
#_______Problema 1____________

# apartado c)

def get_VEF(df2, sexo, fumador):
    gender= (df2['Sexo'])
    smoker= (df2['Fumador'])
    fev= (df2['VEF'])
    age= (df2['Edad'])
    lista = []
    for i in range(len(df2)):
        g = gender[i]
        s = smoker[i]
        f = fev[i]
        a = age[i]
        if g == int(sexo) and s == int(fumador):
            if a>=10 and a<=14:
                lista.append(f)
    return (lista)
    
```

```python
get_VEF(df,sexo = 1,fumador = 0)
```

```python
#obtenemos las listas de los grupos objetivo
chicos_fumadores = get_VEF(df,1,1)
chicos_no_fumadores = get_VEF(df,1,0)
chicas_fumadoras = get_VEF(df,0,1)
chicas_no_fumadoras = get_VEF(df,0,0)
```

```python
plt.hist(chicos_fumadores)
plt.title('Chicos fumadores')
plt.show()
```
